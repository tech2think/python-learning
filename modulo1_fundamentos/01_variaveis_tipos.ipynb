Neste notebook vamos aprender os tipos de dados fundamentais do Python usando exemplos do setor de distribuição de energia elétrica.

1. Tipos primitivos: `int`, `float`, `str`, `bool`

4. Exemplo prático: representando um consumidor como dicionário Python

---

> **Contexto de negócio:** Toda análise de dados começa com a compreensão de como os dados são estruturados. No setor elétrico, trabalhamos com consumidores (pessoas físicas ou jurídicas), suas unidades consumidoras (UCs), medições e histórico de faturamento.

## 1. Inteiros (`int`)

Usamos `int` para códigos identificadores, tensões nominais e contagens.

```python
# Códigos e identificadores
cod_consumidor = 100042
cod_unidade_consumidora = 2001588
num_medidor = 40019273

# Tensões nominais (em Volts)
tensao_monofasico = 220
tensao_trifasico = 380
tensao_media_tensao = 13800  # 13,8 kV

# Contagem de consumidores
total_consumidores_municipio = 45320

print(f"Código UC: {cod_unidade_consumidora}")
print(f"Tipo: {type(cod_unidade_consumidora)}")
print(f"Tensão monofásica: {tensao_monofasico} V")
print(f"Tensão trifásica: {tensao_trifasico} V")
```

## 2. Números de Ponto Flutuante (`float`)

Usamos `float` para medições físicas: consumo em kWh, potência em kW, fator de potência, etc.

```python
# Medições de energia
consumo_mes_kwh = 342.7          # consumo mensal em kWh
potencia_instalada_kw = 15.5     # potência instalada
demanda_registrada_kw = 12.3     # demanda registrada no medidor
fator_potencia = 0.92            # adimensional (0 a 1)

# Tarifas (R$/kWh — valores fictícios para exemplo)
tarifa_b1_kwh = 0.8512           # tarifa residencial em R$/kWh
tarifa_b3_kwh = 0.7234           # tarifa comercial

# Cálculo de fatura básico
valor_fatura = consumo_mes_kwh * tarifa_b1_kwh
print(f"Consumo: {consumo_mes_kwh} kWh")
print(f"Tarifa B1: R$ {tarifa_b1_kwh:.4f}/kWh")
print(f"Valor da fatura: R$ {valor_fatura:.2f}")
print(f"Tipo de consumo_mes_kwh: {type(consumo_mes_kwh)}")
```

## 3. Strings (`str`)

Usamos `str` para nomes, endereços, CPF/CNPJ, códigos alfanuméricos e identificadores de categoria.

```python
# Dados cadastrais
nom_consumidor = "Maria da Silva Santos"
cpf = "123.456.789-00"
cnpj = "12.345.678/0001-90"

# Endereço
nom_logradouro = "Rua das Acácias"
num_logradouro = "245"
nom_bairro = "Jardim das Flores"
nom_municipio = "São Paulo"
cod_uf = "SP"
num_cep = "01310-100"

# Modalidade tarifária e classe de consumo
# B = Baixa Tensão; A = Alta Tensão
cod_modalidade = "B1"  # Residencial baixa tensão
cod_classe = "RESIDENCIAL"

# Operações com strings
nome_upper = nom_consumidor.upper()
nome_lower = nom_consumidor.lower()
nome_titulo = nom_consumidor.title()

print(f"Nome original:  {nom_consumidor}")
print(f"Maiúsculas:     {nome_upper}")
print(f"Minúsculas:     {nome_lower}")
print(f"Title case:     {nome_titulo}")

# Endereço formatado
endereco_completo = f"{nom_logradouro}, {num_logradouro} — {nom_bairro}, {nom_municipio}/{cod_uf} — CEP {num_cep}"
print(f"\nEndereço: {endereco_completo}")
```

## 4. Booleanos (`bool`)

Usados para flags de status: ativo/inativo, inadimplente/regular, baixa renda, etc.

```python
# Flags de status do consumidor
flg_ativo = True
flg_inadimplente = False
flg_baixa_renda = True
flg_geracao_distribuida = False  # possui mini/microgeração?

# Operações lógicas
elegivel_corte = flg_inadimplente and flg_ativo
requer_atencao = flg_inadimplente or flg_baixa_renda

print(f"Consumidor ativo: {flg_ativo}")
print(f"Inadimplente: {flg_inadimplente}")
print(f"Elegível para corte: {elegivel_corte}")
print(f"Requer atenção: {requer_atencao}")

# Booleanos em contexto numérico (útil para somar contagens)
consumidores_status = [True, True, False, True, False, True]
total_ativos = sum(consumidores_status)
print(f"\nTotal de ativos: {total_ativos} de {len(consumidores_status)}")
```

## 5. Listas (`list`)

Listas armazenam sequências de valores — perfeitas para séries históricas de leituras.

```python
# Histórico de leituras mensais (kWh) — últimos 12 meses
leituras_mensais = [310, 325, 342, 298, 315, 401, 389, 356, 303, 295, 318, 342]

# Meses correspondentes
meses = ["Jan", "Fev", "Mar", "Abr", "Mai", "Jun",
         "Jul", "Ago", "Set", "Out", "Nov", "Dez"]

# Estatísticas básicas
media_anual = sum(leituras_mensais) / len(leituras_mensais)
consumo_maximo = max(leituras_mensais)
consumo_minimo = min(leituras_mensais)
mes_maximo = meses[leituras_mensais.index(consumo_maximo)]

print(f"Leituras: {leituras_mensais}")
print(f"Média anual: {media_anual:.1f} kWh/mês")
print(f"Máximo: {consumo_maximo} kWh ({mes_maximo})")
print(f"Mínimo: {consumo_minimo} kWh")

# Adicionando nova leitura e removendo a mais antiga
nova_leitura = 355
leituras_mensais.append(nova_leitura)
leituras_mensais.pop(0)  # remove a primeira
print(f"\nApós atualização: {leituras_mensais}")
```

## 6. Dicionários (`dict`)

Dicionários são a estrutura mais importante para representar registros estruturados — como um consumidor com todos os seus atributos.

```python
# Registro completo de uma Unidade Consumidora
consumidor = {
    "cod_consumidor": 100042,
    "nom_consumidor": "Maria da Silva Santos",
    "cpf_cnpj": "123.456.789-00",
    "cod_modalidade_tarifaria": "B1",
    "cod_classe_consumo": "RESIDENCIAL",
    "nom_logradouro": "Rua das Acácias",
    "num_logradouro": "245",
    "nom_bairro": "Jardim das Flores",
    "nom_municipio": "São Paulo",
    "cod_uf": "SP",
    "num_cep": "01310-100",
    "num_medidor": 40019273,
    "flg_ativo": True,
    "flg_baixa_renda": True,
    "vlr_consumo_medio_kwh": 342.7,
    "leituras_historico_kwh": [310, 325, 342, 298, 315, 401]
}

# Acessando campos
print(f"Consumidor: {consumidor['nom_consumidor']}")
print(f"Código: {consumidor['cod_consumidor']}")
print(f"Classe: {consumidor['cod_classe_consumo']}")
print(f"Município: {consumidor['nom_municipio']}/{consumidor['cod_uf']}")
print(f"Consumo médio: {consumidor['vlr_consumo_medio_kwh']} kWh")

# Verificando chave
if "cpf_cnpj" in consumidor:
    print(f"CPF/CNPJ: {consumidor['cpf_cnpj']}")

# Usando .get() para campos opcionais (evita KeyError)
latitude = consumidor.get("num_latitude", "não informada")
print(f"Latitude: {latitude}")
```

## 7. Datas com `datetime`

Datas são fundamentais no setor elétrico: vigência de contrato, data de leitura, período de faturamento.

```python
from datetime import datetime, date, timedelta

# Data atual
hoje = date.today()
agora = datetime.now()

# Datas do consumidor
dat_inicio_vigencia = date(2018, 3, 15)
dat_ultima_leitura = date(2025, 2, 10)

# Calcular tempo desde última leitura
dias_sem_leitura = (hoje - dat_ultima_leitura).days

# Tempo de cadastro
anos_cadastro = (hoje - dat_inicio_vigencia).days / 365.25

print(f"Data de hoje: {hoje}")
print(f"Início da vigência: {dat_inicio_vigencia}")
print(f"Última leitura: {dat_ultima_leitura}")
print(f"Dias desde última leitura: {dias_sem_leitura}")
print(f"Anos de cadastro: {anos_cadastro:.1f} anos")

# Formatação de datas
data_formatada = dat_ultima_leitura.strftime("%d/%m/%Y")
print(f"\nData formatada (BR): {data_formatada}")

# Parseando string de data
data_string = "15/01/2025"
data_obj = datetime.strptime(data_string, "%d/%m/%Y").date()
print(f"String '{data_string}' → date object: {data_obj}")
```
