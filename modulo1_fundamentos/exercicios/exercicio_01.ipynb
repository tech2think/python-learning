## Soluções

Tente resolver os exercícios antes de ver as soluções!

````python
# SOLUÇÃO — Exercício 1
contagem_classes = {}
for c in carteira:
    classe = c["classe"]
    contagem_classes[classe] = contagem_classes.get(classe, 0) + 1

print("Contagem por classe:")
for classe, qtd in sorted(contagem_classes.items(), key=lambda x: x[1], reverse=True):
    print(f"  {classe:<20}: {qtd}")
````

````python
# SOLUÇÃO — Exercício 2
consumo_por_classe = {}
for c in carteira:
    classe = c["classe"]
    if classe not in consumo_por_classe:
        consumo_por_classe[classe] = []
    consumo_por_classe[classe].append(c["consumo_kwh"])

media_por_classe = {
    classe: round(sum(valores) / len(valores), 1)
    for classe, valores in consumo_por_classe.items()
}

print("Consumo médio por classe (ordenado):")
for classe, media in sorted(media_por_classe.items(), key=lambda x: x[1], reverse=True):
    print(f"  {classe:<20}: {media:>10,.1f} kWh")
````

````python
# SOLUÇÃO — Exercício 3
def classificar_porte(consumo_kwh: float) -> str:
    if consumo_kwh <= 300:
        return "Micro"
    elif consumo_kwh <= 5000:
        return "Pequeno"
    elif consumo_kwh <= 50000:
        return "Médio"
    else:
        return "Grande"

print(f"{'Nome':<25} | {'Consumo (kWh)':>14} | {'Porte'}")
print("-" * 55)
for c in sorted(carteira, key=lambda x: x["consumo_kwh"]):
    porte = classificar_porte(c["consumo_kwh"])
    print(f"{c['nome']:<25} | {c['consumo_kwh']:>14,.1f} | {porte}")
````

````python
# SOLUÇÃO — Exercício 4
problemas_cpf = []

for c in carteira:
    doc = c["cpf_cnpj"]
    
    if doc is None:
        problemas_cpf.append({
            "cod": c["cod"],
            "nome": c["nome"],
            "motivo": "CPF/CNPJ ausente (None)"
        })
        continue
    
    doc_limpo = limpar_documento(doc)
    if len(doc_limpo) not in [11, 14]:
        problemas_cpf.append({
            "cod": c["cod"],
            "nome": c["nome"],
            "motivo": f"Formato inválido: '{doc}' ({len(doc_limpo)} dígitos)"
        })

print(f"Consumidores com problemas no CPF/CNPJ: {len(problemas_cpf)}")
for p in problemas_cpf:
    print(f"  UC {p['cod']}: {p['nome']} — {p['motivo']}")
````
